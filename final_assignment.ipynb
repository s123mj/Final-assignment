```python
#Load Libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline
```

```python
#Load Dataset
dataset = pd.read_csv('survey.csv')
dataset.head()
```

```python
dataset.info()
```

```python
#Treatment Vs Country
#dataset['country_count'] = Counter(dataset['Country'].dropna().tolist()).most_common(10)
fig,ax =plt.subplots(figsize=(14,8))
sns.countplot(data = dataset,x = 'treatment', hue='Country', ax=ax)
plt.title('Treatment Vs Location')
```

```python
dataset['age_Group'] = pd.cut(dataset['Age'].dropna(),
                         [0,18,25,35,45,99],
                         labels=['<18','18-24','25-34','35-44','45+'])
```

```python
fig,ax = plt.subplots(figsize=(14,8))
sns.countplot(data=dataset,x = 'age_Group',hue='Country',ax=ax)
plt.title('Age vs Location')
```
